

##Data Preprocessing
- In this section, prepare the data you have, for training the model.
-Use Label Encoding and One-Hot Encoding techniques to deal with categorical
variables.
- Split your dataset into X_train,X_test, y_train, y_test.
- Scale the dataset by normalizing it(Min-Max Scaling or Standard Scaling).


```python
df.head()
```

```python
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['region'] = label_encoder.fit_transform(df['region'])
df['smoker'] = label_encoder.fit_transform(df['smoker'])
df
```

One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)

one_hot_encoded_data = pd.get_dummies(df, columns = ['region'])
print(one_hot_encoded_data)

df = one_hot_encoded_data
```

```python
df.head()
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))

names = df.columns
d = scaler.fit_transform(df)

scaled_df = pd.DataFrame(d, columns=names)
scaled_df.head()
```

```python
scaled_df.apply(LabelEncoder().fit_transform)
```

```python
# Create the X variable for features
X = scaled_df.loc[:, ['age', 'sex', 'bmi','children', 'smoker', 'region_0', 'region_1', 'region_2', 'region_3']]

# Create the y variable for output labels
y = scaled_df.loc[:, 'charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
y_test
```

##Model Selection

```python
# Create an object
Support_vector_regression_model = SVR()
Tree_model = DecisionTreeRegressor()
Bayesian_model = BayesianRidge()
XGBR_model=XGBRegressor()
RandomForest_model = RandomForestRegressor()
```

```python
# Train
Support_vector_regression_model.fit(X_train, y_train)
Tree_model.fit(X_train, y_train)
Bayesian_model.fit(X_train, y_train)
XGBR_model.fit(X_train, y_train)
RandomForest_model.fit(X_train, y_train)
```

```python
# Make prediction
svr_pred = Support_vector_regression_model.predict(X_test)
tree_pred = Tree_model.predict(X_test)
bayesian_pred = Bayesian_model.predict(X_test)
XG_pred = XGBR_model.predict(X_test)
Rand_pred = RandomForest_model.predict(X_test)
```

```python
svr_report = r2_score(y_test, svr_pred, multioutput='variance_weighted')
tree_report = r2_score(y_test, tree_pred, multioutput='variance_weighted')
bayesian_report = r2_score(y_test, bayesian_pred, multioutput='variance_weighted')
XG_report = r2_score(y_test, XG_pred, multioutput='variance_weighted')
rand_report = r2_score(y_test, Rand_pred, multioutput='variance_weighted')
```

```python
print ('svr_report: '  + str(svr_report))
print ('tree_report: '  + str(tree_report))
print ('bayesian_report: '  + str(bayesian_report))
print ('XG_report: '  + str(XG_report))
print ('rand_report: '  + str(rand_report))
```

##Hyper-parameter Optimization

```python
#Use grid search with the one works the best which is XGBR_model with a score of 0.89
XGBR_model.get_params()
```

```python
model0 = XGBRegressor(max_depth=6, random_state=1, verbosity=1)
model1 = XGBRegressor(max_depth=1, random_state=1, verbosity=1)
model2 = XGBRegressor(max_depth=36, random_state=1, verbosity=1)
model3 = XGBRegressor(max_depth=60, random_state=1, verbosity=1)
```

```python
model0.fit(X_train, y_train)
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
```

```python
pred0 = model0.predict(X_test)
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)
pred3 = model3.predict(X_test)
```

```python
m0_report = r2_score(y_test, pred0, multioutput='variance_weighted')
m1_report = r2_score(y_test, pred1, multioutput='variance_weighted')
m2_report = r2_score(y_test, pred2, multioutput='variance_weighted')
m3_report = r2_score(y_test, pred3, multioutput='variance_weighted')
```

```python
print ('0: '  + str(m0_report))
print ('1: '  + str(m1_report))
print ('2: '  + str(m2_report))
print ('3: '  + str(m3_report))

```
